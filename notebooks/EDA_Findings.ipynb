 EDA Findings Summary

* Mutual fund NAVs showed an overall upward trend from 2022 to 2025 despite market fluctuations.
* SBI Mutual Fund maintained the highest Assets Under Management (AUM) among major fund houses.
* SIP inflows increased steadily over the analysis period, indicating growing investor confidence.
* Flexi Cap and Mid Cap categories attracted strong net inflows compared to other categories.
* Investors aged 26–35 constitute the largest investor segment.
* Investment activity is spread across multiple states, showing increasing mutual fund adoption nationwide.
* Industry folio counts grew significantly, reflecting rising retail investor participation.
* Strong positive correlations exist between 3-year and 5-year fund returns.
* Banking and Pharma sectors account for a significant share of portfolio allocations.
* Most mutual fund schemes maintain expense ratios within acceptable industry standards.
